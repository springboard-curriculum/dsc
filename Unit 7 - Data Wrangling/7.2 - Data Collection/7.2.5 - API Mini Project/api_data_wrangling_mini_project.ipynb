These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
start_date = '2017-01-01'
end_date = '2017-12-01'
url=f'https://data.nasdaq.com/api/v3/datasets/FSE/AFX_X/data.json?start_date={start_date}&end_date={end_date}&order=asc&'
response = requests.get(url)
afx_data_2017 = response.json()
print(json.dumps(afx_data_2017,indent=2))
```

```python
## Extract the column names of the data
column_names = afx_data_2017['dataset_data']['column_names']
print(column_names)
column_names.index('Open')

## Extract the 'data' from the 
afx_trade_data_2017 = afx_data_2017['dataset_data']['data']
## Check data type and sample row
print(type(afx_trade_data_2017))
print(afx_trade_data_2017[0])
```

```python
## Its difficult to keep track of which element in data list represent which attribute. Also accessing this attribute
## using the hard-coded index is error prone and difficult to understand. 
## So we will use the ".index()" on column_names to get the data. Example get the Opening price for first row:
afx_trade_data_2017[0][column_names.index('Open')]

## 3. Caclulate the highest and lowest opening prices
highest_open_price = 0
lowest_open_price = 0
none_count=0
total_count = 0
for row in afx_trade_data_2017:
    if(row[column_names.index('Open')] != None ):
        if(row[column_names.index('Open')] > highest_open_price):
            highest_open_price = row[column_names.index('Open')]
        if(row[column_names.index('Open')] < lowest_open_price or lowest_open_price == 0):
            lowest_open_price = row[column_names.index('Open')]
    else:
        none_count += 1
    total_count += 1

print(f'Total number of trading days : {total_count}')
print(f'Data not available for trading days : {none_count}')
print(f'Highest opening prices for AFX_X in 2017 was : {highest_open_price}')
print(f'Lowest opening prices for AFX_X in 2017 was : {lowest_open_price}')


open_price_list = [x[column_names.index('Open')] for x in afx_trade_data_2017]
## Replace none values with 0
#open_price_list = [0 if x is None else x for x in open_price_list]

## Remove None values from the list
print(len(open_price_list))
open_price_list = [x for x in open_price_list if x]
print(len(open_price_list))

## Sort the list in and take the 1st and last element depending on order of sort. Here its descending 
## Finding highest and lowest can also be done using the max() and min() functions
open_price_list.sort(reverse=True)
print(f'Highest opening prices for AFX_X in 2017 was : {open_price_list[0]}')
print(f'Lowest opening prices for AFX_X in 2017 was : {open_price_list[-1]}')
```

```python
## 4. What was the largest change in any one day (based on High and Low price)?

## Count number of None values for 'Change' 
ctr = sum(row[column_names.index('Change')] is None for row in afx_trade_data_2017)
print(f'None count is {ctr}')

## 236 out of 237 records have Change as None. 
## So we will calculate Change value (Change = High - Low) and store it in change_intra_day
change_intra_day = [round(x[column_names.index('High')] - x[column_names.index('Low')], 2) for x in afx_trade_data_2017]
print(change_intra_day[0:5])
print(f'Largest change in any one day for AFX_X in 2017 was : {max(change_intra_day)}')
```

```python
## 5. What was the largest change between any two days (based on Closing Price)?


## API provides some parameters to fetch the data that is required here. 
##   column_index = 4 : this is Index for Closing price column
##   transform = diff : This calculates row-on-row change i.e. z[t] = y[t]  y[t-1] 
##   order = asc : This is to order the data on date column in ascending order

## concatenate the query parameters to URL
url= url + '&column_index=4&transform=diff&order=asc'
print(f'URL to fetch day change in closing price : {url}\n')
response_change = requests.get(url)

response_change_dict = response_change.json()
afx_day_over_day_change = response_change_dict['dataset_data']['data']
#print(afx_day_over_day_change)

## Sort by column 1 which is change value.
afx_day_over_day_change.sort(key=lambda x:x[1])
## Verify the first 3 and last 3 values
print(afx_day_over_day_change[0:3])
print(afx_day_over_day_change[-3:],'\n')
print(f'Largest change between any two days (based on Closing Price) was {afx_day_over_day_change[-1][1]} on {afx_day_over_day_change[-1][0]}')
```

```python
## 6. What was the average daily trading volume during this year?

## Count number of None values for 'Traded Volume' 
ctr=0
ctr = sum(row[column_names.index('Traded Volume')] is None for row in afx_trade_data_2017)
print(f'None count is {ctr}')

## None count is 0 for 'Traded Volume', so we can directly use the values
## Calculate Average of 'Traded Volume'
avg_traded_volume = sum([x[column_names.index('Traded Volume')] for x in afx_trade_data_2017])/len(afx_trade_data_2017)
avg_traded_volume = round(avg_traded_volume, 2)
print(f'Average daily trading volume during year 2017 was {avg_traded_volume}')

```

```python
## 7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)
def caclulate_median(number_list):
    ## sort the list
    number_list.sort()
    print('Sorted list : ',number_list)
    ## When number of elements is odd, then return quotient; 
    ##    e.g length=101 median will be element 51 or index 50; Quotient of 101/2 is 50
    ## When number of elements is even, then return mean of middle 2 values; 
    median = 0
    if len(number_list)%2 == 1:
        print(f'ODD - Median position is {len(number_list)//2}\n' )
        median = number_list[len(number_list)//2]
    else:        
        median_pos = int(len(number_list)/2)
        print(f'EVEN - median_position is {median_pos-1} and {median_pos}\n')
        median = round(sum(number_list[median_pos-1:median_pos+1])/2,2)
    return median

traded_volume = [x[column_names.index('Traded Volume')] for x in afx_trade_data_2017]
median_traded_volume = caclulate_median(traded_volume)
print(f'Median trading volume during year 2017 was {median_traded_volume}')
```
